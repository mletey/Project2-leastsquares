```julia
using LinearAlgebra 
```

```julia
function ls_interpolate(pmax, npoints, rangestart, rangeend, targetfunc)
    # pmax = number of x coefficients
    # npoints = number of data evaluation points
    # targetfunc = function to be approximated
    
    # evaluation data points
    x = LinRange(rangestart, rangeend, npoints)
    # target value
    b = targetfunc.(x)
    # interpolation matrix
    A = [x[i]^p for i in 1:npoints, p in 0:pmax] # rows are values of discretised x, columns are polynomial exponents

    # Find solution of coefficients C0 to Cp giving approximate power series
    c = (A' * A) \ (A' * b)
    
    return c
end
```

```julia
# Approximate sine with 1000 datapoints between 0 and pi/2
pmax = 10
npoints = 1000
rangestart = 0
rangeend = Ï€/2
targetfunc(x) = sin(x)

coefs = ls_interpolate(pmax, npoints, rangestart, rangeend, targetfunc)
f(coefs,x) = sum(coefs[p+1] * x^p for p in 0:pmax)
```

```julia
using WGLMakie
```

```julia
# graph results
fig = Figure()
ax = Axis(fig[1, 1],
    title = "Approximation on 0 to pi range",
    xlabel = "x",
    ylabel = "y"
)
xx = LinRange(0,Ï€,500)
yy = [f(coefs,x) for x in xx]
ff = [targetfunc.(x) for x in xx]
lines!(xx,yy,label="approximation",linewidth=5)
lines!(xx,ff,label="true sine",linewidth=5)
axislegend()
fig
```

Thus we see the approximation works well in the range 0 to pi, even though it was only fit on the range 0 to pi/2

```julia
# graph results
fig = Figure()
ax = Axis(fig[1, 1],
    title = "Approximation on 0 to 2pi range",
    xlabel = "x",
    ylabel = "y"
)
xx = LinRange(0,2*Ï€,1000)
yy = [f(coefs,x) for x in xx]
ff = [targetfunc.(x) for x in xx]
lines!(xx,yy,label="approximation",linewidth=5)
lines!(xx,ff,label="true sine",linewidth=5)
axislegend()
fig
```
